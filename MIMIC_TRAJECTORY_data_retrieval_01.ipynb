```python
PATH_CG = "/content/drive/MyDrive/NIH/Data/CAREGIVERS.csv"
df_cg = pd.read_csv(PATH_CG)
df_cg.sample(5)
```

#### check other categories

```python
df_notes["CATEGORY"].value_counts()
```

```python
condition = ((df_notes["CATEGORY"]=="Nursing/other") | (df_notes["CATEGORY"]=="Nursing"))
df_notes_nursing = df_notes[condition].copy().reset_index(drop=True)
df_notes_nursing
```

add number of words per each note

```python
def num_words(notes):
  return len(notes.split())

df_notes_nursing['text_len'] = df_notes_nursing['TEXT'].map(lambda x: num_words(x))
```

remove caregivers with no ids

```python
# remove empyt CGID rows 
condition = df_notes_nursing["CGID"].notna()
df_notes_nursing = df_notes_nursing[condition]
df_notes_nursing
```

```python
df_notes_nursing['DESCRIPTION'].value_counts()
```

take only progess notes and remove duplicates

```python
condition = ((df_notes_nursing['DESCRIPTION']=="Report") & (df_notes_nursing['text_len']>10))
cols= ['ROW_ID','SUBJECT_ID','CGID','CHARTDATE','TEXT'] 
df_notes_nursing_filtered = df_notes_nursing[condition].sort_values('text_len', ascending=False).drop_duplicates(['CGID','SUBJECT_ID','CHARTDATE']).reset_index(drop=True)
df_notes_nursing_filtered = df_notes_nursing_filtered[cols]
```

add CGID_LABEL and description

```python
df_notes_nursing_filtered["CGID_LABEL"] = df_notes_nursing_filtered['CGID'].map(lambda x: df_cg[df_cg['CGID']==x]['LABEL'].values[0])
df_notes_nursing_filtered["CGID_DESCRIPTION"] = df_notes_nursing_filtered['CGID'].map(lambda x: df_cg[df_cg['CGID']==x]['DESCRIPTION'].values[0])
df_notes_nursing_filtered
```

```python
df_notes_nursing_filtered['CGID_DESCRIPTION'].value_counts()
```

take only nurses' notes 

```python
condition = (df_notes_nursing_filtered['CGID_DESCRIPTION'] == 'RN') | (df_notes_nursing_filtered['CGID_DESCRIPTION'].isna())
df_notes_nursing_filtered = df_notes_nursing_filtered[condition].reset_index(drop=True)
df_notes_nursing_filtered.set_index(['CGID','SUBJECT_ID','CHARTDATE']).sort_index(ascending=True)['TEXT']
```

```python
# path and name of the cvs file 
PATH_TO_SAVE = "/content/drive/MyDrive/NIH/Data/"
name_file_to_save = 'NOTES_NURSING_TRAJECTORIES_01.csv'
df_notes_nursing_filtered.drop(['CGID_LABEL','CGID_DESCRIPTION'],axis=1).to_csv(PATH_TO_SAVE+name_file_to_save,index=False)
```
